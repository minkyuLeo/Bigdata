### 패배한 경기 중 긍정 댓글 비율이 승리한 경기의 긍정 댓글 비율(0.44) 보다 높은 경우 추출

```python
import pandas as pd
import numpy as np
```

```python
# 업로드된 CSV 파일 불러오기
file_path = '/content/drive/MyDrive/Colab Notebooks/빅데이터/최종프로젝트/all_comments.csv'
df = pd.read_csv(file_path)
```

```python
# Date, Time, Referee를 복합키로 하여 match_id 생성
df['match_id'] = df.groupby(['Date', 'Time', 'Referee']).ngroup() + 1
```

```python
# sentiment가 Positive인 데이터의 comment만 추출
positive_comments = df[df['sentiment'] == 'Positive']['comment'].dropna().tolist()

# 결과 확인 (상위 5개 댓글 출력)
positive_comments[:5]
```

```python
import pandas as pd
import numpy as np
from gensim import corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import re

# NLTK 리소스 다운로드 (stopwords, punkt_tab)
nltk.download('stopwords')
nltk.download('punkt_tab') # Download the 'punkt_tab' resource specifically.

# 텍스트 전처리 함수 정의
def preprocess(text):
    text = text.lower()  # 소문자 변환
    text = re.sub(r'[^a-z\s]', '', text)  # 알파벳과 공백 제외 제거
    tokens = word_tokenize(text)  # 토큰화
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # 불용어 제거
    tokens = [word for word in tokens if len(word) > 2]  # 길이가 2 이하인 단어 제거
    return tokens

# 모든 긍정 댓글 전처리
tokenized_comments = [preprocess(comment) for comment in positive_comments]
```

```python

# 사전(Dictionary) 및 코퍼스(Corpus) 생성
dictionary = corpora.Dictionary(tokenized_comments)
corpus = [dictionary.doc2bow(text) for text in tokenized_comments]

# LDA 모델 학습 (5개의 주제, 15번 반복)
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=15)

# 각 주제별 주요 단어 출력
topics = lda_model.print_topics(num_words=10)
topics
```
