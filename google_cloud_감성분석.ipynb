```python
import os
import json
from google.cloud import language_v1

# 서비스 계정 키 파일 경로 설정
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/your-service-account-file.json"

# Google Cloud Natural Language API 클라이언트 초기화
client = language_v1.LanguageServiceClient()

def analyze_sentiment_with_google(comment):
    """Google Cloud Natural Language API를 사용하여 감성 분석 수행"""
    try:
        # 요청 생성
        document = language_v1.Document(content=comment, type_=language_v1.Document.Type.PLAIN_TEXT)
        response = client.analyze_sentiment(request={"document": document})
        sentiment_score = response.document_sentiment.score

        # 감성 점수에 따라 결과 반환
        if sentiment_score > 0.25:
            return "Positive"
        elif sentiment_score < -0.25:
            return "Negative"
        else:
            return "Neutral"
    except Exception as e:
        print(f"Google Cloud API 호출 중 오류 발생: {comment}. 오류 메시지: {e}")
        return "Error"

# JSON 파일에서 데이터 읽기
with open("cleaned_english_comments_no_emojis.json", "r") as file:
    data = json.load(file)

# 결과 저장할 리스트 초기화
results = []

# 각 경기 데이터를 반복
for match in data:
    comments = match.get("comments", [])
    analyzed_comments = []

    positive_count = 0
    negative_count = 0
    total_comments = len(comments)

    for comment in comments:
        sentiment = analyze_sentiment_with_google(comment)
        analyzed_comments.append({"comment": comment, "sentiment": sentiment})

        if sentiment == "Positive":
            positive_count += 1
        elif sentiment == "Negative":
            negative_count += 1

    positive_ratio = positive_count / total_comments if total_comments > 0 else 0
    negative_ratio = negative_count / total_comments if total_comments > 0 else 0

    match_result = {
        "csv_column": match.get("csv_column"),
        "title": match.get("title"),
        "analyzed_comments": analyzed_comments,
        "sentiment_ratios": {
            "positive_ratio": positive_ratio,
            "negative_ratio": negative_ratio
        }
    }
    results.append(match_result)

# 분석 결과를 JSON 파일로 저장
output_file_name = "comments_with_sentiments.json"
with open(output_file_name, "w") as output_file:
    json.dump(results, output_file, indent=4)

print(f"감성 분석이 완료되었습니다. 결과는 {output_file_name}에 저장되었습니다.")
```
