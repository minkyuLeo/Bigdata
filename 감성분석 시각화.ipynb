## 모듈 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 구글클라우드 감성분석 시각화

## csv파일 불러오기

```python
# CSV 파일 불러오기
df = pd.read_csv('google_sentiment_result.csv')

# 데이터 타입 변환
df["GF"] = pd.to_numeric(df["GF"], errors='coerce')
df["GA"] = pd.to_numeric(df["GA"], errors='coerce')
df["xG"] = pd.to_numeric(df["xG"], errors='coerce')
df["xGA"] = pd.to_numeric(df["xGA"], errors='coerce')
df["Poss"] = pd.to_numeric(df["Poss"], errors='coerce')
df["POSITIVE"] = pd.to_numeric(df["Positive Sentiment Ratio"], errors='coerce')
df["NEGATIVE"] = pd.to_numeric(df["Negative Sentiment Ratio"], errors='coerce')
```

## 데이터 분류

```python
# 결과(Result)에 따른 데이터 분류
df_win = df[df["Result"] == "W"]
df_loss = df[df["Result"] == "L"]
df_draw = df[df["Result"] == "D"]
```

## 시각화

```python
# Negative Sentiment Distribution
plt.figure(figsize=(12, 6))
plt.hist(df_win["NEGATIVE"], bins=20, alpha=0.5, label='Win', edgecolor='black')
plt.hist(df_loss["NEGATIVE"], bins=20, alpha=0.5, label='Loss', edgecolor='black')
plt.hist(df_draw["NEGATIVE"], bins=20, alpha=0.5, label='Draw', edgecolor='black')

# 평균선 추가
if not df_win["NEGATIVE"].empty:
    plt.axvline(df_win["NEGATIVE"].mean(), color='blue', linestyle='dashed', linewidth=2, label=f'Win Mean: {df_win["NEGATIVE"].mean():.2f}')
if not df_loss["NEGATIVE"].empty:
    plt.axvline(df_loss["NEGATIVE"].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Loss Mean: {df_loss["NEGATIVE"].mean():.2f}')
if not df_draw["NEGATIVE"].empty:
    plt.axvline(df_draw["NEGATIVE"].mean(), color='green', linestyle='dashed', linewidth=2, label=f'Draw Mean: {df_draw["NEGATIVE"].mean():.2f}')

plt.title('Negative Sentiment Distribution by Result (with Mean)')
plt.xlabel('NEGATIVE Sentiment')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Positive Sentiment Distribution
plt.figure(figsize=(12, 6))
plt.hist(df_win["POSITIVE"], bins=20, alpha=0.5, label='Win', edgecolor='black')
plt.hist(df_loss["POSITIVE"], bins=20, alpha=0.5, label='Loss', edgecolor='black')
plt.hist(df_draw["POSITIVE"], bins=20, alpha=0.5, label='Draw', edgecolor='black')

# 평균선 추가
if not df_win["POSITIVE"].empty:
    plt.axvline(df_win["POSITIVE"].mean(), color='blue', linestyle='dashed', linewidth=2, label=f'Win Mean: {df_win["POSITIVE"].mean():.2f}')
if not df_loss["POSITIVE"].empty:
    plt.axvline(df_loss["POSITIVE"].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Loss Mean: {df_loss["POSITIVE"].mean():.2f}')
if not df_draw["POSITIVE"].empty:
    plt.axvline(df_draw["POSITIVE"].mean(), color='green', linestyle='dashed', linewidth=2, label=f'Draw Mean: {df_draw["POSITIVE"].mean():.2f}')

plt.title('Positive Sentiment Distribution by Result (with Mean)')
plt.xlabel('POSITIVE Sentiment')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
