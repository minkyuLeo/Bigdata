감성분석을 통한 경기의 반응지수를 경기스탯과 합쳐서 csv로 만들기

```python
import pandas as pd
```

```python
import json

# Load the JSON file to extract all match data
file_path = 'sentiment_analysis_with_original_structure.json'
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extract relevant match data and sentiment ratios
matches_data = []
for match in data:
    csv_data = match.get("csv_column", {})
    sentiment_ratios = match.get('comments_analysis', {}).get('sentiment_ratios', {})
     
    matches_data.append({
        'Date': csv_data.get('Date'),
        'Result': csv_data.get('Result'),
        'GF': csv_data.get('GF'),
        'GA': csv_data.get('GA'),
        'xG': csv_data.get('xG'),
        'xGA': csv_data.get('xGA'),
        'Poss': csv_data.get('Poss'),
        'Opponent': csv_data.get('Opponent'),
        'Venue': csv_data.get('Venue'),
        'Positive Sentiment Ratio': sentiment_ratios.get('POSITIVE'),
        'Negative Sentiment Ratio': sentiment_ratios.get('NEGATIVE')
    })

# Create a DataFrame from the extracted data
df_all_matches = pd.DataFrame(matches_data)

# Display the complete DataFrame
df_all_matches
```

```python
df_all_matches.to_csv("bert_sentiment_result.csv",index=False)
```
