### 패배한 경기 중 긍정 댓글 비율이 승리한 경기의 긍정 댓글 비율(0.44) 보다 높은 경우 추출

```python
import pandas as pd
import numpy as np
```

```python
# 업로드된 CSV 파일 불러오기
file_path = '/content/drive/MyDrive/Colab Notebooks/빅데이터/최종프로젝트/all_comments.csv'
df = pd.read_csv(file_path)
```

```python
# Date, Time, Referee를 복합키로 하여 match_id 생성
df['match_id'] = df.groupby(['Date', 'Time', 'Referee']).ngroup() + 1
```

```python
df
```

```python

# 'Result'가 'L'이고 'positive_ratio'가 0.44보다 높은 데이터 필터링
filtered_df = df[(df['Result'] == 'L') & (df['positive_ratio'] > 0.44)]

filtered_df
```

```python
# 리셋 인덱스
filtered_df=filtered_df.reset_index(drop=True)
filtered_df
```

```python
filtered_df['match_id'].value_counts()
```

총 5경기 이다.

```python
# 'gf'에서 'ga'를 뺀 값을 새로운 열 'gf_minus_ga'에 저장합니다.
filtered_df['GF'] = pd.to_numeric(filtered_df['GF'], errors='coerce') # 'GF' 열을 숫자형으로 변환합니다. errors='coerce'는 변환할 수 없는 값을 NaN으로 처리합니다.
filtered_df['GA'] = pd.to_numeric(filtered_df['GA'], errors='coerce') # 'GA' 열을 숫자형으로 변환합니다. errors='coerce'는 변환할 수 없는 값을 NaN으로 처리합니다.
filtered_df['gf_minus_ga'] = filtered_df['GF'] - filtered_df['GA']

# 'gf_minus_ga' 열의 유니크한 값의 개수를 출력합니다.
print(filtered_df['gf_minus_ga'].nunique())
```

```python
# 1점차, 2점차, 4점차 경기인 match_id의 고유값 추출
one_point_matches = filtered_df[filtered_df['gf_minus_ga'] == -1]['match_id'].unique()
two_point_matches = filtered_df[filtered_df['gf_minus_ga'] == -2]['match_id'].unique()
four_point_matches = filtered_df[filtered_df['gf_minus_ga'] == -4]['match_id'].unique()

print("1점차 경기 match_id:", one_point_matches)
print("2점차 경기 match_id:", two_point_matches)
print("4점차 경기 match_id:", four_point_matches)
```

## 댓글만 추출하여 전처리

```python
#filtered_df의 comment만 추출하여 저장
comments = filtered_df['comment'].tolist()
```

```python
#927개의 댓글 추출
len(comments)
```

```python
#comments에 빈 문자열 제거
comments = [comment for comment in comments if comment != ""] #빈 문자열 제거
print(len(comments))
```

```python
#comments 의 각 개체의 타입을 구해서 타입별 갯수 구하기
type_counts = {}
for comment in comments:
    comment_type = type(comment)
    if comment_type not in type_counts:
        type_counts[comment_type] = 0
    type_counts[comment_type] += 1

for comment_type, count in type_counts.items():
    print(f"Type: {comment_type}, Count: {count}")
```

```python
# comments 리스트에서 float 자료형 제거
comments = [comment for comment in comments if not isinstance(comment, float) and comment] # Remove np.isnan check as it's not applicable to strings

# 소문자화
comments = [comment.lower() for comment in comments]
```

```python
import contractions

# 축약어 처리
expanded_comments = []
for comment in comments:
    expanded_comments.append(contractions.fix(comment))

comments = expanded_comments
```

```python
#토큰화하기
import nltk
from nltk.tokenize import word_tokenize

# 필요한 NLTK 데이터 다운로드
# 'punkt' 대신 'punkt_tab'를 다운로드합니다.
nltk.download('punkt_tab')

# 토큰화
tokenized_comments = []
for comment in comments:
    tokens = word_tokenize(comment)  # word_tokenize를 사용하여 토큰화
    tokenized_comments.append(tokens)

# 토큰화된 결과 출력
for tokens in tokenized_comments:
     print(tokens)
```

```python
# 불용어 제거

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# 불용어 제거
filtered_tokens = []
for tokens in tokenized_comments:
    filtered_tokens.append([w for w in tokens if not w in stop_words])
```

```python
# 불용어 제거된 토큰 출력
for tokens in filtered_tokens:
    print(tokens)
```

```python
# prompt: tokens에서 영문자만 남기기

import re

# 영문자만 남기는 함수 정의
def keep_english_letters(tokens):
    return [re.sub(r'[^a-zA-Z]', '', token) for token in tokens]

# 불용어 제거된 토큰에 함수 적용
filtered_tokens_english_only = []
for tokens in filtered_tokens:
    filtered_tokens_english_only.append(keep_english_letters(tokens))

# 결과 출력
for tokens in filtered_tokens_english_only:
    print(tokens)
```

```python
# 제거되고 남은 빈칸 삭제

# 빈 문자열 제거
filtered_tokens_english_only = [[token for token in sublist if token] for sublist in filtered_tokens_english_only]

# 결과 출력 (빈 문자열 제거 후)
for tokens in filtered_tokens_english_only:
    print(tokens)
```

```python
# 모든 토큰을 하나의 리스트로 합칩니다.
all_tokens = [token for sublist in filtered_tokens_english_only for token in sublist]
```

```python
#5506개의 토큰 존제
len(all_tokens)
```

```python
# 가장 많이 나온 단어에는 어떤 것들이 있는지 barplot을 그려 확인

from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# 단어 빈도 계산
word_counts = Counter(all_tokens)

# 상위 20개 단어 추출
top_20_words = word_counts.most_common(20)

# 단어와 빈도 분리
words, frequencies = zip(*top_20_words)

# barplot 생성
plt.figure(figsize=(12, 6))
sns.barplot(x=frequencies, y=words)
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Top 20 Frequent Words')
plt.show()
```

```python
# 의미를 파악하기 힘든 epl 팀명 제거

# epl_team 리스트
epl_teams = [
    "Arsenal", "Aston Villa", "Bournemouth", "Brentford",
    "Brighton", "Chelsea", "Crystal Palace", "Everton",
    "Fulham", "Leeds", "Leicester", "Liverpool",
    "Manchester City", "Manchester United", "Newcastle United",
    "Nottingham Forest", "Southampton", "Tottenham Hotspur",
    "West Ham United", "Wolverhampton Wanderers"
]

# 리스트 내포(list comprehension)를 사용하여 소문자로 변환
epl_teams = [team.lower() for team in epl_teams]

print(epl_teams)
```

```python
# 띄어쓰기가 있는 팀명을 쪼개서 epl_teams에 추가
for team in epl_teams[:]:  # 원본 리스트를 복사하여 반복
    if " " in team:
        split_teams = team.split(" ")
        epl_teams.extend(split_teams)  # 쪼갠 팀명을 epl_teams에 추가

print(epl_teams)
```

```python
# epl 팀명 제거
all_tokens_del_teams = [token for token in all_tokens if token not in epl_teams]
```
